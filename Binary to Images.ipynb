```python
import os
import numpy as np
from PIL import Image

# Define the function to convert and save the image
def convertAndSave(array, name, root):
    print('Processing ' + name)
    if array.shape[1] != 16:
        raise ValueError("Input array should have 16 columns.")
    b = int((array.shape[0] * 16) ** (0.5))
    b = 2 ** (int(np.log2(b)) + 1)
    a = int(array.shape[0] * 16 / b)
    array = array[:a * b // 16, :]
    array = np.reshape(array, (a, b))
    im = Image.fromarray(np.uint8(array))
    im.save(os.path.join(root, name + '.png'), "PNG")
    return im

# Set the root directory containing the ".bytes" files
root = r"/home/shinesh/Documents/dataSample/Gatak"

# List all files in the root directory
file_names = os.listdir(root)

# Loop through each file
for file_name in file_names:
    # Check if the file has the ".bytes" extension
    if file_name.endswith(".bytes"):
        # Remove the extension to get the file name
        name = os.path.splitext(file_name)[0]

        # Read the contents of the file
        file_path = os.path.join(root, file_name)
        with open(file_path, "r") as f:
            array = []
            for line in f:
                xx = line.split()
                if len(xx) == 17:
                    array.append([int(i, 16) if i != "??" else 0 for i in xx[1:]])

        # Convert and save the image
        convertAndSave(np.array(array), name, root)
```
