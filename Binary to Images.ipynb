```python
import os
import numpy as np
from PIL import Image

# Define the function to convert and save the image
def convertAndSave(array, name, root):
    print('Processing ' + name)
    if array.shape[1] != 16:
        raise ValueError("Input array should have 16 columns.")
    b = int((array.shape[0] * 16) ** (0.5))
    b = 2 ** (int(np.log2(b)) + 1)
    a = int(array.shape[0] * 16 / b)
    array = array[:a * b // 16, :]
    array = np.reshape(array, (a, b))
    im = Image.fromarray(np.uint8(array))
    im.save(os.path.join(root, name + '.png'), "PNG")
    return im

# Set the root directory containing the ".bytes" file
root = r"C:\Users\hp\Downloads\dataSample (1)"

# Specify the name of the ".bytes" file (without the extension)
name = "0A32eTdBKayjCWhZqDOQ"

# Read the contents of the ".bytes" file
file_path = os.path.join(root, name + ".bytes")
with open(file_path, "r") as f:
    array = []
    for line in f:
        xx = line.split()
        if len(xx) == 17:
            array.append([int(i, 16) if i != "??" else 0 for i in xx[1:]])

# Convert and save the image
convertAndSave(np.array(array), name, root)
```
